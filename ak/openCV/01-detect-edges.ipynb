## 3. Edge Detection

The tighter the threshold, the lesser the number of edges.

The wider the threshold, the more number of edges.

```python
# Apply Canny
canny = cv2.Canny(img, threshold1=100, threshold2=210)
```

```python
# Apply blur
blur = cv2.GaussianBlur(img, (5,5), 0)
# Apply Canny
blur_canny = cv2.Canny(blur, threshold1=100, threshold2=210)
```

```python
# Apply Gray
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# Apply blur
blur = cv2.GaussianBlur(gray, (5,5), 0)
# Apply Canny
gray_blur_canny = cv2.Canny(blur, threshold1=100, threshold2=210)
```

```python
# view image in OpenCV
cv2.imshow("Frame View", canny)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 4. Export Edges

```python
cv2.imwrite("./data/images/edges.jpg", canny) # just canny
cv2.imwrite("./data/images/edges_blur_canny.jpg", blur_canny) # blur and canny
cv2.imwrite("./data/images/edges_gray_blur_canny.jpg", gray_blur_canny) # gray, blur and canny
```
